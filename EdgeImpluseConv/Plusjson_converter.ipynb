```python
import json
import os

# Load the JSON files
with open('ei-cuphop-train-y.json', 'r') as file1, open('infolabels.json', 'r') as file2:
    data1 = json.load(file1)
    data2 = json.load(file2)

# Label name to ID mapping
label_mapping = {"Furyo": 1, "Ibutsu": 2, "cup": 3}

# Prepare new data
new_data = []

for sample1, sample2 in zip(data1['samples'], data2['files']):
    sample_id = sample1['sampleId']
    
    # Remove prefix and suffix
    filename_with_ext = os.path.basename(sample2['path'])  # get the filename with extension
    filename = filename_with_ext.split('-')[1]  # remove prefix
    sample_path = filename.split('.png')[0] + '.png'  # keep only until .png

    bounding_boxes = []
    for bbox in sample2['boundingBoxes']:
        # Convert label names to IDs and adjust box coordinates
        label = label_mapping[bbox['label']]
        bounding_boxes.append({
            "label": label,
            "x": bbox['x'],
            "y": bbox['y'],
            "w": bbox['width'],
            "h": bbox['height']
        })

    new_data.append({
        "sampleId": sample_id,
        "path": sample_path,
        "boundingBoxes": bounding_boxes
    })

# Write new data to a JSON file
with open('new_data.json', 'w') as file:
    json.dump({"version": 1, "samples": new_data}, file, indent=4)
```
